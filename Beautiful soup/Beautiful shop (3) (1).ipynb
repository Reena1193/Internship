1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data
https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
def IMDB_movies(url):
    URL = "https://www.imdb.com/list/ls056092300/" 

    r = requests.get(url) 
    soup = BeautifulSoup(r.content, 'html5lib')
    table = soup.find_all('div', {'class': 'lister-item-content'})

    names = []
    ratings = []
    years = []
    for movie in table:
        name = movie.find('a').get_text()
        names.append(name)

        rating = movie.find('span', {'class': 'ipl-rating-star__rating'}).get_text()
        ratings.append(float(rating))

        year = movie.find('span', {'class': 'lister-item-year text-muted unbold'}).get_text()
        years.append(year[1:-1])

    df = pd.DataFrame({'Name': names, 'Rating': ratings, 'Year': years})

    return df

IMDB_movies(URL)
```

7) Please visit https://www.cnbc.com/world/?region=world and scrapa) headings
    b) date
    c) News link

```python
def news_heading(url):
    resaponse = requests.get(url)

    soup = BeautifulSoup(resaponse.content, 'html.parser')

    divs = soup.find_all('div', {'class': 'SidebarArticle-noSticky'})

    headings = []
    dates = []
    links = []

    for div in divs:
        heading = div.find('a')['title']
        headings.append(heading)

        date = div.find('time',{'class':"LatestNews-timestamp"}).get_text()
        dates.append(date)

        link = div.find('a')['href']
        links.append(link)

    df = pd.DataFrame({'Heading': headings, 'Date': dates, 'Link': links})
    
    return df
URL = "https://www.cnbc.com/world/?region=world"
news_heading(URL)
```

 8) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap-
 a) Paper title
 b) date
 c) Author

```python
def agriculture(url):
    
    resaponse = requests.get(url)

    soup = BeautifulSoup(resaponse.content, 'html.parser')

    divs = soup.find_all('section', {'class': 'panel'})

    paper_titles = []
    authors = []

    for div in divs:
        paper_title = div.find('a').get_text()
        paper_titles.append(paper_title)

        author = div.find('p').get_text()
        authors.append(author)

    df = pd.DataFrame({'Paper_title': paper_titles, 'Authors': authors})

    return df

url =  'https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/'
agriculture(url)
```
